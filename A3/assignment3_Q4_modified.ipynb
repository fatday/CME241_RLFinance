```python
'''
fill in the code below, and then run the code in the next cell for the simulation
'''
# integral
import scipy.integrate as integrate
import numpy as np
# simulation parameters
p = 5  # Cost of understocking (per unit)
h = 2  # Cost of overstocking (per unit)
lambda_param = 1.0  # Parameter for exponential distribution

# Probability density function for demand
def demand_pdf(x):
    return lambda_param * np.e **(-lambda_param * x)

# Cumulative distribution function for demand
def demand_cdf(x):
    return 1 - np.e**(-lambda_param*x)

# Expected cost function g(S)
def expected_cost(S):
    # g1(S): Understocking cost
    g1 = p * integrate.quad(lambda x: (x - S) * demand_pdf(x), S, np.inf)[0]

    # g2(S): Overstocking cost (integral using CDF)
    g2 = h * integrate.quad(lambda x: demand_cdf(x), 0, S)[0]

    return g1 + g2
```

```python
result = minimize_scalar(expected_cost, bounds=(0, 10), method='bounded')
optimal_S = result.x

# Simulation of costs
S_values = np.linspace(0, 10, 500)
costs = [expected_cost(S) for S in S_values]

# Plotting the costs against S
plt.figure(figsize=(8, 5))
plt.plot(S_values, costs, label="Expected Cost $g(S)$", color='blue')
plt.axvline(optimal_S, color='red', linestyle='--', label=f"Optimal $S^* \\approx {optimal_S:.2f}$")
plt.xlabel("Supply $S$")
plt.ylabel("Expected Cost $g(S)$")
plt.title("Expected Cost $g(S)$ vs. Supply $S$ (Exponential Demand)")
plt.legend()
plt.grid(True)
plt.show()

optimal_S
```

### Part (B) Answer

```python
import numpy as np
from itertools import product
from scipy.stats import poisson

H = 10
W = 30
alpha = 0.08
beta = 0.82
gamma = 0.95

states = list(range(1, W + 1))
actions = [(l, s) for l in range(H + 1) for s in range(H - l + 1)]

def transition_prob(w, l, s, w_next):
    if w == W:
        return 1.0 if w_next == W else 0.0

    if w_next == W:
        if w == W-1:
            return (1- poisson.cdf(0, mu=alpha * l)) * (1 - beta * s / H) + (beta * s / H)
        else:
            return 1 - poisson.cdf(W-w-1, mu=alpha * l)
    poisson_prob = poisson.pmf(w_next - w, mu=alpha * l) if w_next >= w else 0

    if w_next == w+1:
        return (poisson.pmf(1, mu=alpha * l) * (1 - beta * s / H)) + (poisson.cdf(1, mu=alpha * l) * (beta * s / H))
    elif w_next == w:
        return poisson.pmf(0, mu=alpha * l) * (1 - beta * s / H)
    elif w_next > w+ 1:
        return poisson_prob
    else:
        return 0

def reward(w, l, s):
    return w * (H - l - s)

def mdp_model():
    model = {}
    for w in states:
        model[w] = {}
        for (l_act, s_act) in actions:
            transitions = {}
            for w_next in states:
                prob = transition_prob(w, l_act, s_act, w_next)
                if prob > 0:
                    transitions[w_next] = prob
            model[w][(l_act, s_act)] = transitions
    return model

mdp = mdp_model()

print("MDP Model Constructed with", len(states), "states and", len(actions), "actions.")

```

### Part (C) Answer

```python
# fill in with Python code
def value_iteration(mdp, gamma=0.95, theta=1e-7):
    V = {w: 0 for w in states}
    policy = {w: actions[0] for w in states}

    while True:
        delta = 0
        for w in states:
            v = V[w]
            best_action = None
            best_value = float('-inf')

            for (l, s) in actions:
                expected_value = 0
                if (l, s) in mdp[w]:
                    for w_next, prob in mdp[w][(l, s)].items():
                        expected_value += prob * (reward(w, l, s) + gamma * V[w_next])
                    if expected_value > best_value:
                        best_value = expected_value
                        best_action = (l, s)

            V[w] = best_value
            policy[w] = best_action
            delta = max(delta, abs(v - best_value))
        if delta < theta:
            break

    return policy, V

optimal_policy, optimal_values = value_iteration(mdp, 0.95)

print("Optimal Policy Computed.")
print(optimal_values)
```

\begin{align*}
V_\theta(x)&=U(\theta\cdot x)+\beta \cdot  V_\theta((1-\theta)x)\\

& = \frac{(\theta x)^{1-\gamma}}{1-\gamma}+\beta \cdot \left(U(\theta(1-\theta)x) +\beta V_\theta((1-\theta)^2x) \right)\\

```python
import numpy as np
from typing import Dict
from rl.distribution import Categorical
from rl.markov_process import FiniteMarkovRewardProcess, TransitionStep, NonTerminal

def td_lambda_prediction(
    mrp: FiniteMarkovRewardProcess[int],
    gamma: float,
    lambda_: float,
    alpha: float,
    num_episodes: int,
    discount_factor: float = 0.95
) -> Dict[NonTerminal[int], float]:
    """TD(λ) prediction algorithm for tabular case.

    Args:
        mrp: Finite Markov Reward Process
        gamma: Discount factor for future rewards
        lambda_: Lambda parameter for eligibility traces
        alpha: Learning rate for updates
        num_episodes: Number of episodes to run
        discount_factor: Additional discount factor for rewards (if applicable)

    Returns:
        Dictionary mapping states to their estimated values
    """
    # Initialize value function for all states
    V = {NonTerminal(s): 0.0 for s in [0, 1, 2]}

    # Run episodes
    for _ in range(num_episodes):
        # Initialize eligibility traces at the start of each episode
        E = {NonTerminal(s): 0.0 for s in [0, 1, 2]}

        # Generate episode (starting from state 0)
        state = NonTerminal(0)  # Start from state 0
        step = 0

        while isinstance(state, NonTerminal):  # Until reaching a terminal state
            # Sample next state and reward from the transition model
            next_dist = mrp.transition_reward(state)
            next_state, reward = next_dist.sample()

            # Compute the TD error
            if isinstance(next_state, NonTerminal):  # Non-terminal next state
                delta = reward + gamma * V[next_state] - V[state]
            else:  # Terminal next state
                delta = reward - V[state]  # No future reward from terminal state

            # Update eligibility traces for all states
            E[state] += 1  # Increase eligibility of current state
            for s in V.keys():
                # Update value function using TD(λ) formula
                V[s] += alpha * delta * E[s]
                E[s] *= gamma * lambda_  # Decay eligibility traces

            # Move to the next state
            state = next_state
            step += 1

    return V

# Create a simple test MRP
def get_simple_test_mrp() -> FiniteMarkovRewardProcess[int]:
    '''
    Creates a simple 3-state MRP where:
    - State 0 transitions to 0 or 1 with probabilities 0.3 and 0.7 respectively
    - State 1 transitions to 1 or 2 with probabilities 0.4 and 0.6 respectively
    - State 2 is terminal
    '''
    transition_map: Dict[int, Categorical[Tuple[int, float]]] = {
        0: Categorical({
            (0, 0.0): 0.3,
            (1, 1.0): 0.7
        }),
        1: Categorical({
            (1, 0.0): 0.4,
            (2, 2.0): 0.6
        })
    }
    return FiniteMarkovRewardProcess(transition_map)

# Test the implementation
test_mrp = get_simple_test_mrp()
gamma = 0.9
lambda_ = 0.8
alpha = 0.1
num_episodes = 1000

V = td_lambda_prediction(
    mrp=test_mrp,
    gamma=gamma,
    lambda_=lambda_,
    alpha=alpha,
    num_episodes=num_episodes
)

# Print results
print("Final value function:")
for state, value in V.items():
    print(f"State {state.state}: {value:.3f}")
```

```python
class RandomWalkMRP2D(FiniteMarkovRewardProcess[int]):
    barrier: Tuple[int]
    p: Tuple[float]

    def __init__(
        self,
        barrier: Tuple[int],
        p: Tuple[float]
    ):
        self.barrier = barrier
        self.p = p
        super().__init__(self.get_transition_map())

    def get_transition_map(self) -> Mapping[Tuple[int], Categorical[Tuple[Tuple[int], float]]]:
        '''
        fill in the code below to define the dictionary, d, and then run the code in the next cell for the visualization
        '''
        d: Dict[Tuple[int], Categorical[Tuple[Tuple[int], float]]] = {}

        # For each non-terminal state
        for i in range(1, self.barrier[0]):
            for j in range(1, self.barrier[1]):
                next_states = {}

                # UP movement
                if j < self.barrier[1] - 1:
                    next_states[(i, j+1)] = self.p[0]
                else:  # Terminal state with reward 1
                    next_states[(i, self.barrier[1])] = self.p[0]

                # DOWN movement
                if j > 1:
                    next_states[(i, j-1)] = self.p[1]
                else:  # Terminal state with reward 0
                    next_states[(i, 0)] = self.p[1]

                # LEFT movement
                if i > 1:
                    next_states[(i-1, j)] = self.p[2]
                else:  # Terminal state with reward 0
                    next_states[(0, j)] = self.p[2]

                # RIGHT movement
                if i < self.barrier[0] - 1:
                    next_states[(i+1, j)] = self.p[3]
                else:  # Terminal state with reward 1
                    next_states[(self.barrier[0], j)] = self.p[3]

                # Create distribution for this state
                d[(i,j)] = Categorical({(s, 1.0 if (s[0] == self.barrier[0] or s[1] == self.barrier[1]) else 0.0): p
                                      for s, p in next_states.items()})
        return d
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

from dataclasses import dataclass
from collections import defaultdict
from typing import Mapping, Iterator, TypeVar, Tuple, Dict, Iterable, Generic

# RL imports (adapt or remove if you don't have the same environment):
from rl.distribution import (
    Distribution, Constant, Gaussian, Choose, SampledDistribution, Categorical
)
from rl.markov_process import NonTerminal, State, Terminal
from rl.markov_decision_process import MarkovDecisionProcess
from rl.td import epsilon_greedy_action

S = TypeVar('S')
A = TypeVar('A')
```

```python
# -----------------------------------------------------------------------
# Tabular Q-value function approximation (done for you)
# -----------------------------------------------------------------------
class TabularQValueFunctionApprox(Generic[S, A]):
    """
    A basic implementation of a tabular function approximation
    with constant learning rate of 0.1
    Also tracks the number of updates per (state, action).
    """
    def __init__(self):
        self.counts: Mapping[Tuple[NonTerminal[S], A], int] = defaultdict(int)
        self.values: Mapping[Tuple[NonTerminal[S], A], float] = defaultdict(float)

    def update(self, k: Tuple[NonTerminal[S], A], target: float) -> None:
        alpha = 0.1
        old_val = self.values[k]
        self.values[k] = (1 - alpha) * old_val + alpha * target
        self.counts[k] += 1

    def __call__(self, x: Tuple[NonTerminal[S], A]) -> float:
        return self.values[x]
```

```python
# -----------------------------------------------------------------------
# Double Q-Learning (fill in)
# -----------------------------------------------------------------------
def double_q_learning(
    mdp: MarkovDecisionProcess[S, A],
    states: Distribution[NonTerminal[S]],
    gamma: float
) -> Iterator[TabularQValueFunctionApprox[S, A]]:
    """
    Implements Double Q-Learning as described:
      1) We keep two Q-tables, Q1 and Q2.
      2) We choose actions epsilon-greedily with respect to Q1+Q2.
      3) With 50% chance we update Q1 using next-action chosen by max of Q2,
         otherwise update Q2 using next-action chosen by max of Q1.
    """
    q1 = TabularQValueFunctionApprox()
    q2 = TabularQValueFunctionApprox()

    while True:
        state = states.sample()

        while isinstance(state, NonTerminal):
            # ... rest of the episode logic stays the same ...
            actions = set(mdp.actions(state))
            q_vals = [(q1((state, a)) + q2((state, a))) for a in actions]
            action = list(actions)[np.argmax(q_vals)] if np.random.random() > 0.1 else random.choice(list(actions))

            next_dist = mdp.step(state, action)
            next_state, reward = next_dist.sample()

            if isinstance(next_state, Terminal):
                target = reward
                q1.update((state, action), target)
                q2.update((state, action), target)
            else:
                next_actions = list(mdp.actions(next_state))
                if np.random.random() < 0.5:
                    q2_vals = [q2((next_state, a)) for a in next_actions]
                    best_action = next_actions[np.argmax(q2_vals)]
                    target = reward + gamma * q1((next_state, best_action))
                    q1.update((state, action), target)
                else:
                    q1_vals = [q1((next_state, a)) for a in next_actions]
                    best_action = next_actions[np.argmax(q1_vals)]
                    target = reward + gamma * q2((next_state, best_action))
                    q2.update((state, action), target)

            state = next_state

        # Create and yield average Q function after each episode
        avg_q = TabularQValueFunctionApprox()
        avg_q.values = {k: (q1.values[k] + q2.values[k])/2 for k in q1.values.keys()}
        avg_q.counts = {k: (q1.counts[k] + q2.counts[k])/2 for k in q1.counts.keys()}
        yield avg_q
```

```python
# -----------------------------------------------------------------------
# Standard Q-Learning (fill in)
# -----------------------------------------------------------------------
def q_learning(
  mdp: MarkovDecisionProcess[S, A],
  states: Distribution[NonTerminal[S]],
  gamma: float
) -> Iterator[TabularQValueFunctionApprox[S, A]]:
  """
  Standard Q-Learning:
    1) Keep one Q table
    2) Epsilon-greedy wrt that table
    3) Update Q((s,a)) with  r + gamma * max_{a'} Q((s_next, a'))
  """
  q = TabularQValueFunctionApprox()
  while True:
    state = states.sample()

    while isinstance(state, NonTerminal):
        # Get available actions and their Q-values
        actions = list(mdp.actions(state))  # Convert to list once
        q_vals = [q((state, a)) for a in actions]

        # Choose action using epsilon-greedy
        action = actions[np.argmax(q_vals)] if np.random.random() > 0.1 else random.choice(actions)

        # Take action and observe next state and reward
        next_dist = mdp.step(state, action)
        next_state, reward = next_dist.sample()

        if isinstance(next_state, Terminal):
            target = reward
        else:
            # Get Q-values for next state actions
            next_actions = list(mdp.actions(next_state))

            next_q_vals = [q((next_state, a)) for a in next_actions]
            # Target is reward plus discounted max Q-value of next state
            target = reward + gamma * max(next_q_vals)

        # Update Q-value for current state-action pair
        q.update((state, action), target)
        state = next_state
    yield q
```

```python
# -----------------------------------------------------------------------
# The MDP: States A,B and actions a1,a2,b1,...,bn (don't modify anything anymore, just run to get the graphs)
# -----------------------------------------------------------------------
@dataclass(frozen=True)
class P1State:
    """
    The MDP state, storing whether we are in "A" or "B".
    """
    name: str

class P1MDP(MarkovDecisionProcess[P1State, str]):

    def __init__(self, n: int):
        self.n = n

    def actions(self, state: NonTerminal[P1State]) -> Iterable[str]:
        """
        Return the actions available from this state.
          - if state is A => ["a1", "a2"]
          - if state is B => ["b1", ..., "bn"]
        """
        if state.state.name == "A":
            return ["a1", "a2"]
        else:
            return [f"b{i}" for i in range(1, self.n+1)]

    def step(
        self,
        state: NonTerminal[P1State],
        action: str
    ) -> Distribution[Tuple[State[P1State], float]]:
        """
        Return the distribution of (next state, reward) from (state, action):
          - A + a1 => reward 0, next state B
          - A + a2 => reward 0, next state terminal
          - B + b_i => reward ~ Normal(-0.1,1), next state terminal
        """
        if state.state.name == "A":
            if action == "a1":
                return Constant((NonTerminal(P1State("B")), 0.0))
            else:
                return Constant((Terminal(P1State("T")), 0.0))
        else:
            # For B + b_i => reward ~ N(-0.1,1), then terminal
            def sampler():
                r = np.random.normal(loc=-0.1, scale=1.0)
                return (Terminal(P1State("T")), r)
            return SampledDistribution(sampler)

def run_double_q_once(mdp, start_dist, gamma=1.0, episodes=400):
    """
    Runs one 'chain' of Double Q-Learning for 'episodes' episodes,
    returning a list of Q-values for Q((A,a1)) at the end of each episode.
    """
    dq_iter = double_q_learning(mdp, start_dist, gamma)  # generator
    vals = []
    for _ in range(episodes):
        Q1 = next(dq_iter)
        # record Q1((A,a1)) each time
        qA1 = Q1((NonTerminal(P1State("A")), "a1"))
        vals.append(qA1)
    return vals

def run_q_once(mdp, start_dist, gamma=1.0, episodes=400):
    """
    Same but for standard Q-Learning
    """
    q_iter = q_learning(mdp, start_dist, gamma)
    vals = []
    for _ in range(episodes):
        Q = next(q_iter)
        qA1 = Q((NonTerminal(P1State("A")), "a1"))
        vals.append(qA1)
    return vals

def main():
    # For reproducibility
    random.seed(0)
    np.random.seed(0)

    n = 10
    mdp = P1MDP(n)
    # Always start in A, as a NonTerminal
    start_dist = Constant(NonTerminal(P1State("A")))

    N_RUNS = 100
    N_EPISODES = 400

    all_dbl = []
    all_std = []

    for _ in range(N_RUNS):
        dbl_vals = run_double_q_once(mdp, start_dist, gamma=1.0, episodes=N_EPISODES)
        std_vals = run_q_once(mdp, start_dist, gamma=1.0, episodes=N_EPISODES)
        all_dbl.append(dbl_vals)
        all_std.append(std_vals)

    arr_dbl = np.array(all_dbl)
    arr_std = np.array(all_std)

    avg_dbl = np.mean(arr_dbl, axis=0)
    avg_std = np.mean(arr_std, axis=0)

    plt.figure(figsize=(8,6))
    plt.plot(avg_dbl, label='Double Q-Learning: Q(A,a1)')
    plt.plot(avg_std, label='Q-Learning: Q(A,a1)')
    plt.xlabel('Episode')
    plt.ylabel('Estimated Q-value')
    plt.title('Average Q(A,a1) over 100 runs, n=10')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```

### Part (A) Answer

For **States**, let $U$ be the set of all dices, we define

$$
S=\{(H,T)\mid H:\text{set of dices on hand}\ , \ T:\text{set of dices on table}\}
$$

with $U=H\cup T$ and $H\cap T=\emptyset$ and the terminal state is $(U,\emptyset)$.


For **Actions**, given state $(H,T)$ we consider

$$
A(H,T)=\mathcal{P}(T)\setminus \{\emptyset\}
$$

where $\mathcal{P}(T)$ is the power set of $T$, so this is the set of dices on table we want to take to the hand (table is non-empty).


For **Rewards**, since we only calculate the score when the game ends, but if $1$s on hand larger than $C$ we can calculate the reward earlier, so

$$
R(H,T, a)=
\begin{cases}
\sum_{d\in H\cup a} d&\text{if $\sum_{d\in H\cup a} \mathbb{I}_{d=1}\geq C$ and $U=H\cup a$}\\
0&\text{otherwise}
\end{cases}
$$

where $\mathbb{I}_{d=1}$ is the indicator of a dice has face value $1$.


For **State-Transition Probabilities**, each time we pick an action $a\in \mathcal{P}(T)\setminus \{\emptyset\}$, the table will have $|T\setminus a|$ dices. Since dice has $K$ faces, so it has $K^{|T\setminus a|}$ possible outcomes on the table, then the transition prob will be

$$
P((H,T), a, (H',T'))=
\begin{cases}
\frac{1}{K^{|T\setminus a|}}&\text{if $a\in \mathcal{P}(T)$}\\
0&\text{otherwise}
\end{cases}
$$

where $H'=H\cup a$ and $T'=T\setminus a$.


For **Discount Factor**, since we can only get reward when game end, and there is other information about decays, so the discount factor $\gamma=1$

```python
from itertools import product


class DiceRollGame:

  def __init__(self, N, K, C, gamma = 1):
    self.N = N
    self.K = K
    self.C = C
    self.gamma = gamma
    self.init_state_H = NonTerminal(tuple())
    self.groups = self.generate_H_T_candidates(N, K)
    self.DiceMDP = None
    self.opt_vf = None
    self.opt_policy = None

  def generate_possible_act(self, T):
    return [comb for r in range(1, len(T) + 1) for comb in combinations(T, r)]

  def generate_dice_rolls(self, N, K):
      return [tuple(sorted(roll)) for roll in product(range(1, K + 1), repeat=N)]

  def generate_H_T_candidates(self, N, K):
      groups = {i: self.generate_dice_rolls(i, K) for i in range(1, N + 1)}
      return groups

  def generate_Action_Next_States(self, current_State):
      N, C, K = self.N, self.C, self.K
      cur_state_H, cur_state_T = current_State[0].state, current_State[1].state
      current_hand_size = len(cur_state_H)
      cur_S_counter = {i: 0 for i in range(1, K + 1)}
      cur_S_sum = sum(cur_state_H)
      for x in cur_state_H:
          cur_S_counter[x] += 1

      possible_picktohand_num = N - current_hand_size
      Action_Next_State_Reward = {}

      actions = self.generate_possible_act(cur_state_T)
      for act in actions:
        reward = 0
        if cur_S_counter[1] + act.count(1) >= C and len(act) + len(cur_state_H) == N:
          reward = cur_S_sum + sum(act)
        new_table_size = len(cur_state_T) - len(act)
        next_state_H = tuple(sorted(act + cur_state_H))
        next_state_H = NonTerminal(next_state_H) if len(next_state_H) < N else Terminal(next_state_H)
        Action_Next_State_Reward[NonTerminal(act)] = []
        if new_table_size == 0:
          next_state_T = Terminal(tuple())
          next_state = (next_state_H, next_state_T)
          Action_Next_State_Reward[NonTerminal(act)].append(((next_state, reward), 1.0))
        else:
          next_state_Ts = self.groups[new_table_size]
          probs = 1/len(next_state_Ts)
          for next_state_T in next_state_Ts:
            next_state = (next_state_H, NonTerminal(next_state_T))
            Action_Next_State_Reward[NonTerminal(act)].append(((next_state, reward), probs))

      return Action_Next_State_Reward

  def generate_state_mapping(self):
      mapping = {}
      N, C, K = self.N, self.C, self.K
      for i in self.groups:
        H_states = self.groups[i]
        if i != N:
          T_states = self.groups[N - i]
          for H in H_states:
            for T in T_states:
              mapping[(NonTerminal(H), NonTerminal(T))] = self.generate_Action_Next_States((NonTerminal(H), NonTerminal(T)))
      for T in self.groups[N]:
        mapping[(self.init_state_H, NonTerminal(T))] = self.generate_Action_Next_States((self.init_state_H, NonTerminal(T)))
      return mapping


  def get_DiceMDP(self):
      if self.DiceMDP is not None:
        return self.DiceMDP
      mapping = self.generate_state_mapping()
      self.DiceMDP = FiniteMarkovDecisionProcess(mapping)
      return self.DiceMDP

  def get_opt_v_p(self):
    if self.opt_vf is not None and self.opt_policy is not None:
      return self.opt_vf, self.opt_policy
    self.opt_vf, self.opt_policy  = value_iteration_result(self.DiceMDP, self.gamma)
    return self.opt_vf, self.opt_policy
```
